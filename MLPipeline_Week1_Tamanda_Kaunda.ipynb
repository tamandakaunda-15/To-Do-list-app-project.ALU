```python
# Importing libraries
import pandas as pd
import numpy as np
import requests
from io import StringIO
```

```python
""" Loading the data """

# Read the data
url = "https://pynative.com/wp-content/uploads/2019/01/Automobile_data.csv"

# Send request with headers
headers = {'User-Agent': 'Mozilla/5.0'}
response = requests.get(url, headers=headers)

# Load into pandas
df = pd.read_csv(StringIO(response.text))
df.info()

```

```python
# Print the first and last five columns
print(df.head(5))
print(df.tail(5))
```

```python
# Identify missing values and choose a method to replace them
df_fillvalues = df.replace(['?', 'n.a'], np.nan)

df_fillvalues.head()
```

```python
# Print the most expensive car, the company name, and the price.
expensive_cars = df_fillvalues.loc[df_fillvalues['price'].idxmax()]
print(expensive_cars[['company', 'price']])
```

```python
# Count total cars per company
total_cars_per_compnay = df_fillvalues['company'].value_counts()
print(total_cars_per_compnay)
```

```python
# Find each company’s highest-priced car
higher_price_car = df_fillvalues.groupby('company')['price'].max()
print(higher_price_car)
```

```python
# Find the average mileage of each car-making company
average_mileage_find = df_fillvalues.groupby('company')['average-mileage'].mean()
print(average_mileage_find)
```

```python
# Sort all cars by the price column
sorted_cars = df_fillvalues.sort_values('price', ascending=True)
sorted_cars.head()
```
